```python
# Interactive Audit for AI Governance Lab

import pandas as pd
from deepchecks.tabular import Dataset
from deepchecks.tabular.suites import data_integrity

# 1. Load Data
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", 
           "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", 
           "hours-per-week", "native-country", "income"]
df = pd.read_csv(url, names=columns, na_values=" ?", skipinitialspace=True)

# 2. Configure Deepchecks Dataset
# Highlight 'sex' and 'race' as sensitive features for future Bias/Fairness checks
ds = Dataset(df, label='income', cat_features=['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country'])

# 3. RUN INTERACTIVE INTEGRITY AUDIT
# This command displays the full, interactive report directly in the notebook cell.
integrity_suite = data_integrity()
integrity_suite.run(ds)
```
